```python
from requests import Request, Session

NAVER_CLIENT_ID = "cdJJWjwbSqpl1v23_5m9"
NAVER_CLIENT_SECRET = "Bs0kvhbAx1"

def get_text_list():
    lists = []
    with open('yesterday.txt', 'r',encoding="utf8") as file:
        contents = file.read()
        lists = contents.split("\n")

    return lists

def save_to_file(list):
    with open('yesterday.translated_ver2.txt', 'w',encoding="utf8") as file:
        file.writelines(list)

def main():
    session = Session()

    headers = {
        'X-Naver-Client-Id': NAVER_CLIENT_ID,
        'X-Naver-Client-Secret': NAVER_CLIENT_SECRET,
    }
    url = 'https://openapi.naver.com/v1/papago/n2mt'
    # 파일에서 읽은 데이터를 리스트에 저장
    lists = get_text_list()
    # list comprehension (List에서 0이 아닌것만 저장 한다.)
    lists = [ i for i in lists if len(i) != 0]

    result_list = []
    for en_text in lists:
        payload = {
            'source': 'en',
            'target': 'ko',
            'text': en_text,
        }
        req = Request('POST', url, data=payload, headers=headers)
        prepared = req.prepare()

        res = session.send(prepared)
        print('send original text : ', en_text)
        ko_text = ''
        
        # try ~ except 처리
        try:
            ko_text = res.json()['message']['result']['translatedText']
        except Exception as err:
            print('에러 발생 ',res.json())
            break

        print('translated to : ', ko_text)
        result_list.append(en_text+ '\n')
        result_list.append(ko_text + '\n')

    save_to_file(result_list)
    print('job completed..')


main()
```
